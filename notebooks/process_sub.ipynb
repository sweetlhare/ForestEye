```python
import pandas as pd
from datetime import datetime

# Функция для преобразования строки в datetime
def parse_date(date_str):
    return datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')

# Загрузка данных
df1 = res
df2 = res_cloud

# Преобразование строковых дат в datetime
for df in [df1, df2]:
    df['date_registration_start'] = df['date_registration_start'].apply(parse_date)
    df['date_registration_end'] = df['date_registration_end'].apply(parse_date)

# Функция для проверки пересечения временных интервалов
def intervals_overlap(start1, end1, start2, end2):
    return (start1 <= end2) and (start2 <= end1)

# Поиск пересечений
intersections = []

for _, row1 in df1.iterrows():
    for _, row2 in df2.iterrows():
        if (row1['name_folder'] == row2['name_folder'] and 
            row1['class'] == row2['class'] and 
            intervals_overlap(row1['date_registration_start'], row1['date_registration_end'],
                              row2['date_registration_start'], row2['date_registration_end'])):
            
            # Определение пересекающегося интервала
            start = max(row1['date_registration_start'], row2['date_registration_start'])
            end = min(row1['date_registration_end'], row2['date_registration_end'])
            
            intersections.append({
                'name_folder': row1['name_folder'],
                'class': row1['class'],
                'date_registration_start': start,
                'date_registration_end': end,
                'count_df1': row1['count'],
                'count_df2': row2['count']
            })

# Создание датафрейма с пересечениями
intersection_df = pd.DataFrame(intersections)
```

```python
intersection_df = intersection_df.rename(columns={'count_df1': 'count'}).drop(columns='count_df2')
intersection_df.to_csv('1/intersection_df.csv', index=False)
```

# make 1 intersection better

```python
intersection_df = pd.read_csv('1/intersection_df.csv')
```

```python
# Функция для замены редких классов
def replace_rare_classes(group):
    class_counts = group['class'].value_counts()
    most_common_class = class_counts.index[0]
    rare_classes = class_counts[class_counts == 1].index
    group.loc[group['class'].isin(rare_classes), 'class'] = most_common_class
    return group

# Применение функции замены редких классов
cleaned_df = intersection_df.groupby('name_folder').apply(replace_rare_classes).reset_index(drop=True)

# Вывод результатов до очистки
print("Распределение классов до очистки:")
print(intersection_df.groupby('name_folder')['class'].value_counts())

# Вывод результатов после очистки
print("\nРаспределение классов после очистки:")
print(cleaned_df.groupby('name_folder')['class'].value_counts())

# Сохранение очищенных результатов в CSV файл
cleaned_df.to_csv('cleaned_intersections.csv', index=False)
print("\nОчищенные результаты сохранены в файл 'cleaned_intersections.csv'")
```

# remove all rare class from intersections 1

```python
intersection_df = pd.read_csv('1/intersection_df.csv')
```

```python
# Функция для удаления редких классов
def remove_rare_classes(group):
    class_counts = group['class'].value_counts()
    rare_classes = class_counts[class_counts == 1].index
    return group[~group['class'].isin(rare_classes)]

# Применение функции удаления редких классов
cleaned_df = intersection_df.groupby('name_folder').apply(remove_rare_classes).reset_index(drop=True)

# Вывод результатов до очистки
print("Распределение классов до очистки:")
print(intersection_df.groupby('name_folder')['class'].value_counts())

# Вывод результатов после очистки
print("\nРаспределение классов после очистки:")
print(cleaned_df.groupby('name_folder')['class'].value_counts())

# Подсчет удаленных записей
removed_count = len(intersection_df) - len(cleaned_df)
print(f"\nУдалено записей: {removed_count}")

# Вывод информации о удаленных классах
removed_classes = intersection_df[~intersection_df.index.isin(cleaned_df.index)]
print("\nУдаленные классы по папкам:")
print(removed_classes.groupby('name_folder')['class'].value_counts())

# Сохранение очищенных результатов в CSV файл
cleaned_df.to_csv('cleaned_intersections_removed_rare.csv', index=False)
print("\nОчищенные результаты сохранены в файл 'cleaned_intersections_removed_rare.csv'")

# Сохранение удаленных записей в отдельный CSV файл
removed_classes.to_csv('removed_rare_classes.csv', index=False)
print("Удаленные записи сохранены в файл 'removed_rare_classes.csv'")
```
